```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('predict_energy_data_train.csv')
data
```

```python
X = data[data.columns[1:25]]
y = data['Energy']
print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=4)
```

```python
print("X_trian",X_train.shape)
print("y_train",y_train.shape)
print("X_test",X_test.shape)
print("y_test",y_test.shape)
```

#### Linear Regression

```python
res = LinearRegression(normalize=True)
res.fit(X_train,y_train)
```

```python
y_pred = res.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('mse',mean_squared_error(y_test,y_pred))

print('r2_score',r2_score(y_test,y_pred))
```

### RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
res_random = RandomForestRegressor(n_estimators = 1000)
res_random.fit(X_train,y_train)
```

```python
y_p = res_random.predict(X_test)
print('mse',mean_squared_error(y_test,y_p))
print('r2_score',r2_score(y_test,y_p))
```

### Extratree classifier

```python
from sklearn.ensemble import ExtraTreesRegressor
res_extra = ExtraTreesRegressor(n_estimators = 1000,max_depth = None)
res_extra.fit(X_train,y_train)
y_prediction = res_extra.predict(X_test)
print('mse',mean_squared_error(y_test,y_prediction))
print('r2', r2_score(y_test,y_prediction))
```
